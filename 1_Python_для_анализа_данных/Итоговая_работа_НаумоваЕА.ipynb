## Обработка пропусков (столбец 'price')

```python
# Найдем среднее и медиану по всему столбцу 'price'
print(df['price'].describe())
```

Медиана: 24.0 <br>
Среднее: 33.33 <br>
Значения сильно отличаются, посколько имеются выбросы. В этом случае для пропущенных значений используется медианное значение. <br>
Но поскольку у нас есть столбец country, а цена может отличаться в зависимости от страны, то вычислим значение среднего и медианы цены по каждой стране.

```python
# Сгруппируем цену постранам и посмотрим данные медианы и среднего цены по каждой стране отдельно
print(df.groupby('country')['price'].agg(['mean', 'median']))
print()
```

Видим, что значение среднего так же больше (иногда в 2 раза) значения медианы. Поэтому будем работать дальше с медианным значением для заполнения пропущенных ячеек столбца price.

```python
# Группировка по странам и нахождение медианы цены, вывод медианы в отдельную переменную
median_price_by_country = df.groupby('country')['price'].median()
median_price_by_country = median_price_by_country.fillna(df['price'].median())  # для Egypt
print(median_price_by_country)
```

В обоих случаях медиана (50%) сильно отклоняется от среднего, поэтому заполнять пропущенные значения следует медианным значением.

```python
# Замена пустых значений (NaN) в столбце price на соответствующие медианы
df['price'].fillna(df['country'].map(median_price_by_country), inplace=True)
df
```

## Найти популярные сорта по региону.

```python
# Группировка по региону (province) и сорту (variety) и нахождение максимального рейтинга для каждой группы
popular_varieties_by_region = df.groupby(['province', 'variety'])['points'].max().reset_index()

# Сортировка по региону и рейтингу в убывающем порядке
popular_varieties_by_region = popular_varieties_by_region.sort_values(['province', 'points'])

# Вывод результата (популярные сорта с наибольшим рейтингом для каждого региона)
print(popular_varieties_by_region)
```

Группируем данные по колонкам 'province' (регион) и 'variety' (сорт вина) и
находим максимальный рейтинг для каждой группы с помощью метода max().
Затем сортируем результаты по региону и рейтингу в убывающем порядке.

Визуализацируем зависимость цены от рейтинга. Используем линейную регрессию из `scikit-learn`, чтобы построить модель, которая предсказывает цену продукта на основе его рейтинга. Затем мы визуализируем полученную линейную регрессию на графике с помощью красной линии. Это позволяет увидеть общий тренд зависимости цены от рейтинга.

#  <a name="5">5. Проверка гипотез </a>

## ГИПОТЕЗА №1
    H0: Средние пользовательские рейтинги красного и белого вина одинаковые.
    H1: Средние пользовательские рейтинги красного и белого вина разные.

Для проверки данной гипотезы можно использовать t-тест для независимых выборок. T-тест позволит определить, есть ли статистически значимая разница между средними значениями рейтингов для красного и белого вина.

```python
# Разделение данных на две группы: красное и белое вино
red_wine_points = df[df["color"] == "red"]["points"]
white_wine_points = df[df["color"] == "white"]["points"]

# Выполнение t-теста
t_statistic, p_value = ttest_ind(red_wine_points, white_wine_points, equal_var=True)

# Уровень значимости
alpha = 0.05

print("T-статистика:", t_statistic)
print("p-значение:", p_value)

# Проверка гипотезы
if p_value < alpha:
    print("Отвергаем нулевую гипотезу. Средние пользовательские рейтинги красного и белого вина разные.")
else:
    print("Не можем отвергнуть нулевую гипотезу. Средние пользовательские рейтинги красного и белого вина одинаковые.")
```


В результате p-value меньше выбранного уровня значимости и мы можем отвергнуть нулевую гипотезу в пользу альтернативной. Это означает, что существует статистически значимая разница между средними рейтингами красного и белого вина, и мы можем принять альтернативную гипотезу H1.

## ГИПОТЕЗА №2
    H0: Средние цены двух популярных сортов вина одинаковые.
    H1: Средние цены двух популярных сортов вина разные.

Для проверки гипотезы о равенстве средних цен двух популярных сортов вина, также можно воспользоваться двухвыборочным t-тестом для независимых выборок и определить, есть ли статистически значимая разница между средними значениями цен для двух популярных сортов вина.

```python
# Определение двух популярных сортов вина
# Возьмем 2 популярных сорта вина "Cabernet Blen" и "Chardonnay"
variety1 = "Cabernet Blend"
variety2 = "Chardonnay"

# Выделение данных для каждого из сортов
variety1_prices = df[df["variety"] == variety1]["price"]
variety2_prices = df[df["variety"] == variety2]["price"]

# Выполнение t-теста
t_statistic, p_value = ttest_ind(variety1_prices, variety2_prices, equal_var=True)

# Уровень значимости
alpha = 0.05

print("T-статистика:", t_statistic)
print("p-значение:", p_value)

# Проверка гипотезы
if p_value < alpha:
    print("Отвергаем нулевую гипотезу. Средние цены двух популярных сортов вина разные.")
else:
    print("Не можем отвергнуть нулевую гипотезу. Средние цены двух популярных сортов вина одинаковые.")
```

P-value меньше выбранного уровня значимости, мы можем отвергнуть нулевую гипотезу в пользу альтернативной, т.е., что средние цены для выбранных популярных сортов вина различаются.
